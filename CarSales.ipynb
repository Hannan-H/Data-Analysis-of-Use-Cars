```python
#install specific version of libraries used in  lab
#! mamba install pandas==1.3.3  -y
#! mamba install numpy=1.21.2 -y

# import pandas library
import pandas as pd
import numpy as np

# Data Source:
# Data source: https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data
# Data type: csv

filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(filepath, header=None)

# show the first 5 rows using dataframe.head() method
#print("The first 5 rows of the dataframe") 
#df.head(5)

# create headers list
# This will create headers for each of the columns
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
#print("headers\n", headers)

df.columns = headers
df.columns

#Used to see the first 10 entries
#df.head(10)

# Replacing the '?' symbol with NaN (Not a Number) so the dropna() can remove missing values
df1=df.replace('?',np.NaN)

# This will drop missing values along the column "price"
# axis = 0  means that the contents along the entire row will be dropped wherever the entity 'price' is found to be NaN
df=df1.dropna(subset=["price"], axis=0)
df.head(20)

#This will find the name of the columns of the dataframe.
print(df.columns)

#This will save the dataset
#df.to_csv("automobile.csv", index=False)


# describe all the columns in "df" 
df.describe(include = "all")

#This is used to describe certain coloumns
#df[['length', 'compression-ratio']].describe()

# look at the info of "df"
#This method prints information about a data frame including the index dtype and columns, non-null values and memory usage.
#df.info()
```
